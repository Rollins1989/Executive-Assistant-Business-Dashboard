```python
import pandas as pd
import numpy as np
import os
from faker import Faker
from random import randint, choice, uniform
from datetime import datetime, timedelta

print("Libraries Loaded Successfully")
```

```python
# Root Project Folder
project_folder = "Executive_Assistant_Business_Dashboard"

# List of folders
folders = [
    "data",
    "reports",
    "charts",
    "output",
    "dashboard"
]

# Create root folder
os.makedirs(project_folder, exist_ok=True)

# Create subfolders
for folder in folders:
    os.makedirs(os.path.join(project_folder, folder), exist_ok=True)

print("Folder Structure Created Successfully!")
```

```python
fake = Faker("en_IN")

print(fake.name())
print(fake.company())
```

```python
meeting_data = []

departments = [
    "HR",
    "Sales",
    "Marketing",
    "Finance",
    "IT",
    "Operations"
]

status = [
    "Completed",
    "Scheduled",
    "Cancelled"
]

for i in range(1, 101):

    meeting = {

        "Meeting_ID": f"MT{i:03}",

        "Meeting_Date":
        datetime.today() + timedelta(days=randint(-30,30)),

        "Meeting_Time":
        f"{randint(9,18)}:{choice(['00','30'])}",

        "Client_Name":
        fake.company(),

        "Department":
        choice(departments),

        "Agenda":
        choice([
            "Project Review",
            "Client Discussion",
            "Sales Pitch",
            "Budget Meeting",
            "Weekly Sync",
            "Vendor Discussion"
        ]),

        "Duration_Minutes":
        randint(30,120),

        "Status":
        choice(status),

        "Location":
        choice([
            "Conference Room A",
            "Conference Room B",
            "Zoom",
            "Google Meet"
        ])

    }

    meeting_data.append(meeting)

meeting_df = pd.DataFrame(meeting_data)

meeting_df.head()
```

```python
meeting_file = os.path.join(
    project_folder,
    "data",
    "meetings.xlsx"
)

meeting_df.to_excel(meeting_file, index=False)

print("Meeting Tracker Saved")
```

```python
priority = [
    "High",
    "Medium",
    "Low"
]

task_status = [
    "Open",
    "In Progress",
    "Completed",
    "Pending"
]

owners = [
    "Kuldeep",
    "Rohan",
    "Anjali",
    "Neha",
    "Rahul"
]

task_data = []

for i in range(1,151):

    task = {

        "Task_ID": f"TSK{i:03}",

        "Task_Name":
        fake.sentence(nb_words=4),

        "Owner":
        choice(owners),

        "Priority":
        choice(priority),

        "Status":
        choice(task_status),

        "Start_Date":
        datetime.today() - timedelta(days=randint(1,20)),

        "Deadline":
        datetime.today() + timedelta(days=randint(1,30))

    }

    task_data.append(task)

task_df = pd.DataFrame(task_data)

task_df.head()
```

```python
task_file = os.path.join(
    project_folder,
    "data",
    "tasks.xlsx"
)

task_df.to_excel(task_file,index=False)

print("Task Tracker Saved")
```
