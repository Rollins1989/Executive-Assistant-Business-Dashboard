```python
import pandas as pd
import numpy as np
import os

from faker import Faker
from random import randint, choice, uniform
from datetime import datetime, timedelta

fake = Faker("en_IN")

project_folder = "Executive_Assistant_Business_Dashboard"
```

```python
employees = []

departments = [
    "HR",
    "Finance",
    "Sales",
    "Marketing",
    "IT",
    "Operations"
]

designations = [
    "Executive Assistant",
    "Manager",
    "Senior Executive",
    "Analyst",
    "Team Lead",
    "Coordinator"
]

for i in range(1, 31):

    employees.append({

        "Employee_ID": f"EMP{i:03}",

        "Employee_Name": fake.name(),

        "Department": choice(departments),

        "Designation": choice(designations),

        "Email": fake.email(),

        "Phone": fake.phone_number()

    })

employee_df = pd.DataFrame(employees)

employee_df.head()

employee_df.to_excel(
    os.path.join(project_folder,
                 "data",
                 "employees.xlsx"),
    index=False
)
```

```python
print("All datasets generated successfully.")
```

```python
# ==========================================
# Generate Follow-up Dataset
# ==========================================

followups = []

priorities = ["High", "Medium", "Low"]
status_list = ["Pending", "Completed", "Overdue"]

for i in range(1, 121):

    followups.append({

        "Followup_ID": f"FUP{i:03}",

        "Person_Name": fake.name(),

        "Company": fake.company(),

        "Department": choice(departments),

        "Followup_Date": datetime.today() - timedelta(days=randint(0,30)),

        "Due_Date": datetime.today() + timedelta(days=randint(-10,20)),

        "Priority": choice(priorities),

        "Status": choice(status_list),

        "Remarks": choice([
            "Call back",
            "Send Proposal",
            "Waiting for Approval",
            "Meeting Scheduled",
            "Email Sent"
        ])

    })

followup_df = pd.DataFrame(followups)

followup_df.head()
```

```python
followup_df.to_excel(
    os.path.join(
        project_folder,
        "data",
        "followups.xlsx"
    ),
    index=False
)

print("followups.xlsx created successfully!")
```

```python
# ==========================================
# Generate Follow-up Dataset
# ==========================================

followups = []

priorities = ["High", "Medium", "Low"]

followup_status = [
    "Pending",
    "Completed",
    "In Progress"
]

remarks = [
    "Waiting for response",
    "Email sent",
    "Call scheduled",
    "Proposal shared",
    "Meeting completed",
    "Need management approval",
    "Quotation pending",
    "Follow-up next week"
]

for i in range(1, 121):

    followups.append({

        "Followup_ID": f"FUP{i:03}",

        "Person_Name": fake.name(),

        "Company": fake.company(),

        "Department": choice(departments),

        "Followup_Date": datetime.today() - timedelta(days=randint(0,30)),

        "Due_Date": datetime.today() + timedelta(days=randint(-10,20)),

        "Priority": choice(priorities),

        "Status": choice(followup_status),

        "Remarks": choice(remarks)

    })

followup_df = pd.DataFrame(followups)

followup_df.head()
```

```python
followup_df.to_excel(
    os.path.join(
        project_folder,
        "data",
        "followups.xlsx"
    ),
    index=False
)

print("✅ followups.xlsx created successfully")
```

```python
# ==========================================
# Generate Vendor Dataset
# ==========================================

vendors = []

vendor_categories = [

    "IT Services",

    "Office Supplies",

    "Consulting",

    "Travel",

    "Logistics",

    "Marketing",

    "Security",

    "Software"

]

payment_status = [

    "Paid",

    "Pending",

    "Partially Paid"

]

for i in range(1,81):

    vendors.append({

        "Vendor_ID":f"VEN{i:03}",

        "Vendor_Name":fake.company(),

        "Category":choice(vendor_categories),

        "Contact_Person":fake.name(),

        "Email":fake.email(),

        "Phone":fake.phone_number(),

        "Payment_Status":choice(payment_status),

        "Contract_Value":round(uniform(25000,500000),2),

        "Last_Meeting":datetime.today()-timedelta(days=randint(1,120))

    })

vendor_df = pd.DataFrame(vendors)

vendor_df.head()
```

```python
vendor_df.to_excel(

    os.path.join(

        project_folder,

        "data",

        "vendors.xlsx"

    ),

    index=False

)

print("✅ vendors.xlsx created successfully")
```

```python
# ==========================================
# Generate Expense Dataset
# ==========================================

expenses = []

expense_categories = [
    "Travel",
    "Hotel",
    "Food",
    "Office Supplies",
    "Software",
    "Internet",
    "Training",
    "Utilities",
    "Marketing",
    "Courier"
]

payment_modes = [
    "Cash",
    "UPI",
    "Credit Card",
    "Bank Transfer"
]

approval_status = [
    "Approved",
    "Pending",
    "Rejected"
]

for i in range(1, 501):

    expenses.append({

        "Expense_ID": f"EXP{i:04}",

        "Expense_Date":
        datetime.today() - timedelta(days=randint(0,365)),

        "Category":
        choice(expense_categories),

        "Department":
        choice(departments),

        "Vendor":
        choice(vendor_df["Vendor_Name"].tolist()),

        "Amount":
        round(uniform(500,50000),2),

        "Payment_Mode":
        choice(payment_modes),

        "Approval_Status":
        choice(approval_status),

        "Approved_By":
        choice(employee_df["Employee_Name"].tolist())

    })

expense_df = pd.DataFrame(expenses)

expense_df.head()
```

```python
expense_df.to_excel(

    os.path.join(

        project_folder,

        "data",

        "expenses.xlsx"

    ),

    index=False

)

print("✅ expenses.xlsx created successfully")
```

```python
print(expense_df.shape)

expense_df.sample(5)
```

```python
# ==========================================
# Generate Calendar Dataset
# ==========================================

calendar = []

start_date = datetime(2026,1,1)

for i in range(365):

    current_date = start_date + timedelta(days=i)

    meeting_hours = randint(0,6)

    focus_hours = randint(1,6)

    travel_hours = randint(0,3)

    leave = choice(["Yes","No"])

    wfh = choice(["Yes","No"])

    total_hours = meeting_hours + focus_hours + travel_hours

    calendar.append({

        "Date":current_date,

        "Day":current_date.strftime("%A"),

        "Meeting_Hours":meeting_hours,

        "Focus_Hours":focus_hours,

        "Travel_Hours":travel_hours,

        "Leave":leave,

        "WFH":wfh,

        "Total_Working_Hours":total_hours

    })

calendar_df = pd.DataFrame(calendar)

calendar_df.head()
```

```python
calendar_df.to_excel(

    os.path.join(

        project_folder,

        "data",

        "calendar.xlsx"

    ),

    index=False

)

print("✅ calendar.xlsx created successfully")
```
