```python
# ============================================
# Notebook 11
# Automated Executive Report
# ============================================

import pandas as pd
import os

from datetime import datetime
from reportlab.platypus import (
    SimpleDocTemplate,
    Table,
    TableStyle,
    Paragraph,
    Spacer
)

from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet

print("Libraries Loaded Successfully")
```

```python
project_folder = "Executive_Assistant_Business_Dashboard"

data_folder = os.path.join(project_folder, "data")

reports_folder = os.path.join(project_folder, "reports")

os.makedirs(reports_folder, exist_ok=True)
```

```python
meeting_df = pd.read_excel(
    os.path.join(data_folder, "meetings.xlsx")
)

task_df = pd.read_excel(
    os.path.join(data_folder, "tasks.xlsx")
)

followup_df = pd.read_excel(
    os.path.join(data_folder, "followups.xlsx")
)

vendor_df = pd.read_excel(
    os.path.join(data_folder, "vendors.xlsx")
)

expense_df = pd.read_excel(
    os.path.join(data_folder, "expenses.xlsx")
)

calendar_df = pd.read_excel(
    os.path.join(data_folder, "calendar.xlsx")
)

print("All datasets loaded successfully.")
```

```python
total_meetings = len(meeting_df)

meeting_hours = round(
    meeting_df["Duration_Minutes"].sum()/60,
    2
)

completed_tasks = (
    task_df["Status"]=="Completed"
).sum()

pending_tasks = (
    task_df["Status"]=="Pending"
).sum()

pending_followups = (
    followup_df["Status"]=="Pending"
).sum()

vendor_count = len(vendor_df)

total_expense = round(
    expense_df["Amount"].sum(),
    2
)

working_hours = calendar_df["Total_Working_Hours"].sum()

focus_hours = calendar_df["Focus_Hours"].sum()

productivity = round(
    (focus_hours/working_hours)*100,
    2
)

utilization = round(
    (working_hours/(365*8))*100,
    2
)
```

```python
report = pd.DataFrame({

"Metric":[

"Total Meetings",

"Meeting Hours",

"Completed Tasks",

"Pending Tasks",

"Pending Followups",

"Vendor Count",

"Total Expense",

"Working Hours",

"Productivity",

"Utilization"

],

"Value":[

total_meetings,

meeting_hours,

completed_tasks,

pending_tasks,

pending_followups,

vendor_count,

total_expense,

working_hours,

productivity,

utilization

]

})

report
```

```python
report.to_excel(

os.path.join(

reports_folder,

"Executive_Report.xlsx"

),

index=False

)

print("Executive_Report.xlsx Created")
```

```python
weekly = pd.DataFrame({

"Metric":[

"Meetings",

"Tasks",

"Followups",

"Expenses"

],

"Value":[

total_meetings,

len(task_df),

len(followup_df),

total_expense

]

})

weekly.to_excel(

os.path.join(

reports_folder,

"Weekly_Report.xlsx"

),

index=False

)

print("Weekly_Report.xlsx Created")
```

```python
monthly = expense_df.copy()

monthly["Expense_Date"] = pd.to_datetime(
    monthly["Expense_Date"]
)

monthly["Month"] = monthly["Expense_Date"].dt.month_name()

monthly_report = monthly.groupby(
    "Month"
)["Amount"].sum().reset_index()

monthly_report.to_excel(

os.path.join(

reports_folder,

"Monthly_Report.xlsx"

),

index=False

)

print("Monthly_Report.xlsx Created")
```

```python
print("="*60)

print("BUSINESS INSIGHTS")

print("="*60)

print(f"• Total Meetings Conducted : {total_meetings}")

print(f"• Pending Tasks : {pending_tasks}")

print(f"• Completed Tasks : {completed_tasks}")

print(f"• Pending Follow-ups : {pending_followups}")

print(f"• Vendors Managed : {vendor_count}")

print(f"• Total Expense : ₹{total_expense:,.2f}")

print(f"• Productivity : {productivity}%")

print(f"• Utilization : {utilization}%")
```

```python
pdf_file = os.path.join(
    reports_folder,
    "Executive_Report.pdf"
)

doc = SimpleDocTemplate(pdf_file)

styles = getSampleStyleSheet()

story = []

story.append(
    Paragraph(
        "Executive Business Report",
        styles["Heading1"]
    )
)

story.append(
    Spacer(1,12)
)

table_data = [
    ["Metric","Value"]
]

for i in range(len(report)):
    table_data.append([
        report.iloc[i,0],
        str(report.iloc[i,1])
    ])

table = Table(table_data)

table.setStyle(TableStyle([

('BACKGROUND',(0,0),(-1,0),colors.grey),

('TEXTCOLOR',(0,0),(-1,0),colors.whitesmoke),

('GRID',(0,0),(-1,-1),1,colors.black),

('BACKGROUND',(0,1),(-1,-1),colors.beige),

('ALIGN',(0,0),(-1,-1),'CENTER')

]))

story.append(table)

doc.build(story)

print("Executive_Report.pdf Created Successfully")
```
